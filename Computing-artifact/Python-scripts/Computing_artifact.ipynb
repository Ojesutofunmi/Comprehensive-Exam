$\textbf{CASE 1 - Transport equation}$

This transport equation is the the advection equation

$$\begin{equation}
\label{eq:01}
\frac{\partial q_L}{\partial t} + u \frac{\partial q_L}{\partial x} = 0
\end{equation}$$






$\textbf{CASE 2 - Advection Diffusion equation}$

The linear model equation of the Navier Stokes is the Advection Diffusion equation

$$\begin{equation}
\frac{\partial q_L}{\partial t} + u \frac{\partial q_L}{\partial x} = \nu \frac{\partial^2 q_L}{\partial x^2}
\end{equation}$$





$\textbf{CASE 3 - Advection Diffusion equation plus disturbance}$

$$\begin{equation}
\frac{\partial q_L}{\partial t} + u \frac{\partial q_L}{\partial x} = \nu \frac{\partial^2 q_L}{\partial x^2} + C_f q_L
\end{equation}$$

$\textbf{Definition of terms}$


- $\textbf{u}$ denote the velocity vector measured in $ms^{-1},$
  
The constant $\nu$ is given by $\frac{\mu_T}{\sigma_T}$, measured in $ m^2 s^{-1},$
  
- $\sigma_T$ the effective Prandtl number of the fully turbulent fluid $ = 0.9.$

- $\mu_T$ is the turbulence kinematic viscosity, measured in $ m^2 s^{-1}$ and it is given by

 $$\mu_T = c_\mu \rho \frac{e^2}{\epsilon},$$

where; 

- $c_\mu$ is an empirical constant - model coefficient for the turbulent viscosity = $0.09 $

- $e$ is the instantaneous turbulence kinetic energy $ m^2 s^{-2}$, 

- $\rho$ is the density of air $kg/m^3$

- $\epsilon$ is the dissipation rate $ m^2 s^{-3}$,

 $$ e = \frac{3}{2} (\textbf{u} I)^2,$$


- $I$ is the initial turbulence intensity in $[\%]$ 
  $$ I = 0.16 Re^{-\frac{1}{8}},$$
 
- $Re$ is the turbulent Reynold's number $= 10^6$ 

$$ \epsilon = c_{\mu}^{\frac{3}{4}} e^{\frac{3}{2}}l^{-1},$$

- $l$ is the turbulence or eddy length scale

The turbulent length scale in meters can be estimated as
$${l=0.07L,}$$

- $L$  is a characteristic length measured in meters,

- $q_L$ is the humidity (non-vapor part),  measured by $kg_{water} kg_{air}^{-1},$

- $\textbf{w}_{f}$ is the free-fall velocity of a particle expressed in $ms^{-1}$, 

***



***

- $C_f$ can be called the intensity, that depends on the definition of $\textbf{w}_{f}$.

The following conditions hold for every values of $C_f$;
- if $C_f > 0$ i.e positive - upward movement of snow to to the ground,
- if $C_f < 0$ i.e negative - downward movement of snow to the ground,

```python
#To obtain the value of nu - viscosity term
sigma_t = 0.9
c_mu = 0.09
rho = 1.225
u = 2
L = 5
Re = 10e6
I = 0.16*(Re**(-0.125))
e = 1.5*((u*I)**2)
l = 0.07*L
epsilon = ((c_mu)**(0.75))*(e**1.5)*(l**(-1))
mu_t = c_mu*rho*(e**2 / epsilon)
nu = mu_t / sigma_t 
nu
```

The $1D$ advection equation in flux form is


$\begin{equation}
\frac{\partial q}{\partial t} + \frac{\partial f}{\partial x} = 0 
\end{equation}, \quad$ where $f = qu$



 $ \textbf {Transformation in the Garlekin Method Machinery}$
- multiply the equation by the test function
- integrate over the domain $\Omega_e$


$$\begin{equation}
\int_{\Omega_e} \psi_i \left(\frac{\partial q_{N}^{(e)}}{\partial t} +  \frac{\partial f_{N}^{(e)}}{\partial x}\right) d \Omega_e = \textbf{0 }
\end{equation}$$


where $\psi$ is the basis function with $i = 0, \cdot N,$ and $\textbf{0 }$ is the zero vector with indices $i,j = 0,\cdots,N$

Expanding the variables inside the element and using nodal approximation for N.

$$q_{N}^{(e)} (x,t) = \sum_{j = 0}^{N} q_{j}^{(e)} (t) \psi_j(x), \quad  f_{N}^{(e)} (x,t) = f(q_{N}^{(e)}) =  u q_{N}^{(e)}$$

where $f_N^{(e)} = \sum_{j = 0}^{N} f_j^{(e)} (t) \psi_j(x)$

$$\begin{equation}
\int_{\Omega_e} \psi_i \sum_{j = 0}^{N} \psi_j \frac{d q_{j}^{(e)}}{d t} d \Omega_e  + \int_{\Omega_e}  \psi_i  \sum_{j = 0}^{N} \frac{d \psi_j }{d x}  f_{j}^{(e)} d \Omega_e = \textbf{0 }
\end{equation}$$


The matrix form is written as 

$$\begin{equation}
M_{ij}^{(e)} \frac{dq^{(e)}}{dt}  +   D_{ij}^{(e)}  f_{j}^{(e)} = \textbf{0 }
\end{equation}$$

where the mass matrix is given as $$M_{ij}^{(e)} = \int_{\Omega_e} \psi_i \psi_j d \Omega_e ,$$ and the differentiation matrix $$ D_{ij}^{(e)} = \int_{\Omega_e} \psi_i \frac {d \psi_j} {dx} d \Omega_e $$


- DSS form

$$\begin{equation}
M_{IJ} \frac{dq_J}{dt}  +    D_{IJ}  f_{J} = \textbf{0 }
\end{equation}$$
where $I,J = 1, \cdots, N_P, \qquad N_P$ is the number of points.

$\textbf{Helpful Relations}$

- {Error Norm}

The normalized L2 error norm used is
$$
||error||_{L^2} = \sqrt{ \frac{ \sum_{k=1}^{N_P}  \left( q^{numerical} - q^{exact}(x_k) \right)^2}{ \sum_{k=1}^{N_P} q^{exact}(x_k) ^2} }
$$
where $k=1,...,N_P$ are $N_p=N_e N + 1 $ global gridpoints and $q^{numerical}$ and $q^{exact}$ are the numerical and exact solutions after one  full revolution of the wave.

- {Time-Integrator}

The time-integrator is Runge-Kutta time-integrator which solves the equation

$\frac{\partial q}{\partial t} = RHS(q)$

in two steps:


$q^{n+1/2} = q^{n} + \frac{\Delta t}{2} RHS (q^n)$

$q^{n+1} = q^{n} + \frac{\Delta t}{2} RHS (q^{n+1/2})$

The right-hand-side term can be expressed as a matrix-vector operation $RHS(q) = Rq$, where $R$ is a matrix i derived from the matrix form of the governing equation by rearranging the terms, in the code using mass, differentiation and flux matrices.
